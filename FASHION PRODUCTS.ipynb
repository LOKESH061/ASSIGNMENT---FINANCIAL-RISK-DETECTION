# Label Encoded

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
df.selling_price = lb.fit_transform(df.selling_price)
```

```python
df._id= lb.fit_transform(df._id)
df.actual_price =lb.fit_transform(df.actual_price)
df.average_rating=lb.fit_transform(df.average_rating)
df.brand=lb.fit_transform(df.brand)
```

```python
df.info()
```

```python
df.category=lb.fit_transform(df.category)
df.crawled_at=lb.fit_transform(df. crawled_at)
```

```python
df.description=lb.fit_transform(df.description)
df.discount=lb.fit_transform(df.discount)
```

```python
df.pid=lb.fit_transform(df.pid)
df.seller=lb.fit_transform(df.seller)
```

```python
df.images.value_counts()
```

```python
df.sub_category=lb.fit_transform(df.sub_category)
df.title=lb.fit_transform(df.title)
df.url=lb.fit_transform(df.url)
```

# Data Visualization

* Visualize trends and patterns using various plots

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Popular Product categories:
plt.figure(figsize=(10, 6))
sns.countplot(x='category', data=df, order=df['category'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Count of Products by Category')
plt.show()
```

# Popular Brands

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='brand', data=df, order=df['brand'].value_counts().index[:10])
plt.xticks(rotation=45)
plt.title('Top 10 Brands by Product Count')
plt.show()
```

* Brand0 appears to be the dominant brand in the dataset, accounting for 30000.00% 
of the total products.

* The presence of Brand11,189, 252,195,66 and  Brand 86 among the top brands indicates a 
competitive landscape in the market, with these brands enjoying substantial consumer preference.

# Pricing Trends and Discount

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='crawled_at', y='selling_price', data=df, ci=None)
plt.title('Trend of Selling Price Over Time')
plt.xlabel('Date')
plt.ylabel('Selling Price')
plt.show()
```

* The trend in selling prices appears to exhibit a linear pattern, steadily increasing over the observed time periods.

* This indicates a consistent rise in the prices of fashion products within the dataset, suggesting potential
factors such as inflation, seasonal variations, or changes in market demand.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='discount', data=df)
plt.xticks(rotation=45)
plt.title('Average Discount by Category')
plt.show()
```
